```python
# Dependencies and import data
import pandas as pd
import numpy as py

path = "data/fish_catches.csv"
catches_df = pd.read_csv(path)

catches_df = catches_df[['Species','Country','2006','2007','2008','2009','2010','2011','2012','2013','2014']]

catches_df.head()
```

```python
# Transform data
df_unpivoted = catches_df.melt(id_vars=['Country','Species'], var_name='Year', value_name='TLW')
df_unpivoted
```

```python
# Group by Species and Country for each year 
country_catch = df_unpivoted.groupby(["Country","Species","Year"]).sum()
country_catch.reset_index()
catches_df = pd.DataFrame(country_catch.reset_index())
catches_df.head()
```

```python
# Write to CSV
catches_df.to_csv(r'catches_final.csv', index=False)
```
